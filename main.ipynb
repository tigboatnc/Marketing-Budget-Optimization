# Solving Allocation for Marketing Campaign using CVXPY

```python
import pandas as pd
import cvxpy as cp
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import softmax
```

```python
df_final = pd.read_csv('./dataset.csv')
```

----------------------------------

```python
b = 20000
```

```python
n = 10
```

# 1. Cost Effectiveness 

__Find Most Cost Effective Way to Reach the Biggest Audience__

> __minimize__ c'x  <br/>
> __st.__  Ax <=b

```python
# Picking a small sample 
df_sample = df_final.sample(n=n)

# Viewing the sample 
df_sample
```

```python
# Optimization Algorithm 
np.random.seed(1)

n = len(df_sample)
print('n',n)
x0 = np.random.randn(n)
c = df_sample['follower_count'].to_numpy()
print('c',c)
A = df_sample['CPP'].to_numpy()
print('A',A)

x = cp.Variable(n)
prob = cp.Problem(cp.Maximize(c.T@x),
                 [A.T @ x <= b,x >= 0])

prob.solve()
# Print result.
print("\nThe optimal value is", prob.value)
print("A solution x is")
print(x.value)
print("A dual solution is")
print(prob.constraints[0].dual_value)

```

```python
# Plot to show allocation of budget among different streams
plt.plot(x.value)
plt.show()
```

```python
# Making Observations on the output, Calculating views achieved and allocation
df_sample['Allocation'] = x.value
df_sample['Views'] = df_sample['Allocation'] * df_sample['follower_count']
df_sample['Cost'] = df_sample['Allocation'] * df_sample['CPP']
df_sample
```

```python
print('Total Money Spent in Campaign: ',sum(df_sample['Cost']))
print('Total Views Acquired in Campaign: ',sum(df_sample['Views']))
```
